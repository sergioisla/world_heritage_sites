```python
import pandas as pd
import seaborn as sns

import folium 
from IPython.display import display
import ipywidgets as widgets
from folium import plugins
import branca.colormap as cm
```

```python
# reading UNESCO file
df = pd.read_excel('whc-sites-2019.xlsx')
```

```python
df
```

```python
# exploring "danger_list" column
df.danger_list.unique()
```

```python
# exploring null information
df.isnull().sum()
```

```python
# excluding not useful columns
datamap=df[['latitude','longitude','date_inscribed','states_name_en','name_en']]
```

```python
datamap
```

```python
# Defining the range of values for the colormap
min_value = min(datamap.date_inscribed)
max_value = max(datamap.date_inscribed)

# Creating the colormap
colormap = cm.LinearColormap(['red', 'yellow', 'green', 'blue'],
                             index=[1978, 1990, 2005, 2019], 
                             vmin=min_value, vmax=max_value,
                            caption='Year of inscription')
```

```python
colormap
```

```python
# creating the map

m = folium.Map(location=[0, 0], zoom_start=1)

loc = 'Corpus Christi'
#defining the title
title_html = '''
             <h3 align="center" style="font-size:16px"><b>World Heritage Sites color coded by year of inscription</b></h3>
             '''.format(loc)
#adding the title
m.get_root().html.add_child(folium.Element(title_html))

for index, row in datamap[:].iterrows():
    latitude = row['latitude']
    longitude = row['longitude']
    name = row['states_name_en']
    #color = category_colors.get(category, '#000000')
    year = row['date_inscribed']
    site = row['name_en']
    color = colormap(year)
    
    # Add a marker with a popup for each point
    folium.CircleMarker(
                        location=[latitude, longitude],
                        icon=folium.Icon(color=color),  # Pass the color variable
                        radius=1,
                        color=color,
                        fill=True,
                        fill_opacity=0.6,
                        tooltip=[name,year,site],
                        keep_in_front=True
                        ).add_to(m)
m.add_child(colormap)

m
```
